```python
{
  "cells": [
    {
      "cell_type": "markdown",
      "metadata": {},
      "source": [
        "# 11 - PSD, filter functions, Floquet framing\n",
        "\n",
        "We demonstrate Parseval consistency under TensorFlow FFT conventions and apply idealized filter weights.\n"
      ]
    },
    {
      "cell_type": "code",
      "execution_count": null,
      "metadata": {},
      "outputs": [],
      "source": [
        "import tensorflow as tf\n",
        "from tig.core.random import Rng\n",
        "from tig.fourier.filter_functions import lowpass_ideal, bandpass_ideal\n",
        "\n",
        "tf.keras.backend.set_floatx(\"float64\")\n",
        "tf.config.run_functions_eagerly(True)\n",
        "\n",
        "rng = Rng(seed=0)\n",
        "n = 2048\n",
        "x = rng.normal((n,), dtype=tf.float64)\n",
        "X = tf.signal.fft(tf.cast(x, tf.complex128))\n",
        "\n",
        "lhs = tf.reduce_sum(x * x)\n",
        "rhs = (1.0 / tf.cast(n, tf.float64)) * tf.reduce_sum(tf.cast(tf.abs(X) ** 2, tf.float64))\n",
        "rel = tf.abs(lhs - rhs) / (tf.abs(rhs) + tf.cast(1e-15, tf.float64))\n",
        "print(\"Parseval relative mismatch:\", float(rel.numpy()))\n",
        "\n",
        "omega = tf.linspace(-500.0, 500.0, n)\n",
        "lp = lowpass_ideal(omega, omega_c=120.0)\n",
        "bp = bandpass_ideal(omega, omega_lo=80.0, omega_hi=160.0)\n",
        "\n",
        "print(\"lowpass weight mean:\", float(tf.reduce_mean(lp).numpy()))\n",
        "print(\"bandpass weight mean:\", float(tf.reduce_mean(bp).numpy()))\n"
      ]
    }
  ],
  "metadata": {
    "kernelspec": {
      "display_name": "Python 3",
      "language": "python",
      "name": "python3"
    },
    "language_info": {
      "name": "python",
      "version": "3.x"
    }
  },
  "nbformat": 4,
  "nbformat_minor": 5
}
```
