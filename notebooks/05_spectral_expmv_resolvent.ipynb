```python
{
  "cells": [
    {
      "cell_type": "markdown",
      "metadata": {},
      "source": [
        "# 05 - Spectral: expmv and resolvent\n",
        "\n",
        "We compare `expmv(A,v,t)` to dense `expm(tA)v` on a small instance.\n"
      ]
    },
    {
      "cell_type": "code",
      "execution_count": null,
      "metadata": {},
      "outputs": [],
      "source": [
        "import tensorflow as tf\n",
        "from tig.core.random import Rng\n",
        "from tig.spectral.expmv import expmv\n",
        "\n",
        "tf.keras.backend.set_floatx(\"float64\")\n",
        "tf.config.run_functions_eagerly(True)\n",
        "\n",
        "rng = Rng(seed=0)\n",
        "A = rng.normal((32, 32), dtype=tf.float64)\n",
        "v = rng.normal((32,), dtype=tf.float64)\n",
        "t = 0.25\n",
        "\n",
        "y = expmv(A, v, t=t)\n",
        "y_ref = (tf.linalg.expm(tf.cast(t, tf.float64) * A) @ tf.reshape(v, (-1, 1)))[:, 0]\n",
        "\n",
        "rel = tf.linalg.norm(y - y_ref) / (tf.linalg.norm(y_ref) + tf.cast(1e-15, tf.float64))\n",
        "print(\"relative error:\", float(rel.numpy()))\n"
      ]
    }
  ],
  "metadata": {
    "kernelspec": {
      "display_name": "Python 3",
      "language": "python",
      "name": "python3"
    },
    "language_info": {
      "name": "python",
      "version": "3.x"
    }
  },
  "nbformat": 4,
  "nbformat_minor": 5
}
```
